# Question 1 (40 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

(3) Collect all the reviews of the top 1000 most popular software from G2 or Capterra.

(4) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

(5) Collect all the information of the 904 narrators in the Densho Digital Repository.


```python
# I have taken the movie reviews of Mission Impossible- Dead Recknoing (2023) - Movie Name

import requests
from bs4 import BeautifulSoup
import csv

url_movie = "https://www.imdb.com/title/tt9603212/reviews?ref_=tt_urv"
review_page_no = 42
movie_reviews = []
for review_page in range(1, review_page_no + 1):
    url_review_page = f"{url_movie}?start={((review_page - 1) * 10)}"
    response = requests.get(url_review_page)
    if response.status_code == 200:
        movie_soup = BeautifulSoup(response.text, 'html.parser')
        review_elements = movie_soup.find_all("div", class_="text show-more__control")
        for review_element in review_elements:
            review_text = review_element.get_text(strip=True)
            movie_reviews.append(review_text)
    else:
        print(f"Unable to find reviews in IMDb page {review_page}. Give the working URL.")

with open("Mission_Impossible_movie_reviews.csv", "w", newline="", encoding="utf-8") as csvfile:
    build = csv.writer(csvfile)
    build.writerow(["Viewer Reviews"])
    for viewer_review in movie_reviews:
        build.writerow([viewer_review])

print("I had collected the Movie Reviews of MISSION IMPOSSIBLE-DEAD RECKONING[PART ONE]")
print(f"I had Gathered {len(movie_reviews)} Viewer reviews and stored to 'Mission_Impossible_movie_reviews.csv' file.")
```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

stopwords = stopwords.words("english")
# defining function for cleaning the reviews text data
def text_cleaning(words):
    words = re.sub(r"[^\w\s]", "", words)
    words = re.sub(r"[0-9]", "", words)
    words = " ".join([word for word in words.split() if word not in stopwords])
    words = words.lower()
    #performing stemming
    stemming = PorterStemmer()
    words = " ".join([stemming.stem(word) for word in words.split()])
    #performing lemmatization
    lemmatization = WordNetLemmatizer()
    words = " ".join([lemmatization.lemmatize(word) for word in words.split()])
    return words

with open("Mission_Impossible_movie_reviews.csv", "r") as f:
    access = csv.reader(f)
    next(access)
    with open("Cleaned_Mission_Impossible_movie_reviews.csv", "w", newline="") as f_out:
        build = csv.writer(f_out)
        build.writerow(["Raw_Reviews", "Cleaned_Reviews"])
        for row in access:
            viewer_review = row[0]
            cleaned_review = text_cleaning(viewer_review)
            build.writerow([viewer_review, cleaned_review])

f.close()
f_out.close()
print("Performed stemming.....")
print("Performed lemmatization.....")
print("Done with cleaning the reviews and saved new data to Cleaned_Mission_Impossible_movie_reviews.csv")
```

# Question 3 (30 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
import spacy
eng_model = spacy.load("en_core_web_sm")

def find_POS(words):
    para = eng_model(words)
    types_pos = { 'Adjective': 0, 'Adverb': 0, 'Noun': 0, 'Verb': 0}

    for line in para:

        if line.pos_ == 'ADJ':
            types_pos['Adjective'] += 1
        elif line.pos_ == 'ADV':
            types_pos['Adverb'] += 1
        elif line.pos_ == 'NOUN':
            types_pos['Noun'] += 1
        elif line.pos_ == 'VERB':
            types_pos['Verb'] += 1
    return types_pos

with open("Cleaned_Mission_Impossible_movie_reviews.csv", "r") as f:
    access = csv.reader(f)
    next(access)
    final_pos = { 'Adjective': 0, 'Adverb': 0, 'Noun': 0, 'Verb': 0}

    for row in access:
        new_review = row[1]
        types_pos = find_POS(new_review)
        for pos_reviews, num in types_pos.items():
            final_pos[pos_reviews] += num

print("Final Parts of Speech in data:")
for pos_reviews, num in final_pos.items():
    print(f"{pos_reviews}: {num}")
f.close()
```

```python
import spacy
import csv
eng_model = spacy.load("en_core_web_sm")
with open("Cleaned_Mission_Impossible_movie_reviews.csv", "r") as f:
    access = csv.reader(f)
    next(access)
    count_of_entities = {}

    for line in access:
        new_review = line[1]
        corpus = eng_model(new_review)

        for entity in corpus.ents:
            type_of_entity = entity.label_
            text_entity = entity.text

            if type_of_entity in count_of_entities:
                count_of_entities[type_of_entity].append(text_entity)
            else:
                count_of_entities[type_of_entity] = [text_entity]

quant_of_entity = {type_of_entity: len(entities) for type_of_entity, entities in count_of_entities.items()}

print("Here is the Quantities of all the entities:")
for type_of_entity, quant in quant_of_entity.items():
    print(f"{type_of_entity}: {quant}")

f.close()
```
